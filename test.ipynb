```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 500
X = np.random.randint(0, 100, size= n)
X
```

```python
Y = np.random.random(size=n)*X + np.random.random(size=n)
```

```python
plt.scatter(X, Y)
```

```python
a = sum(X**2)
b = sum(X)
c = b**2
d = sum(X*Y)
e = sum(Y)
```

```python
m = 1/(a-c/n) * ( d- b*e/n)
p = 1/n*( e- m*b)
```

```python
Y_chapeau = m*X + p
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X.reshape(-1, 1), Y)
Y_pred = model.predict(X.reshape(-1, 1))
```
